```python
import pandas as pd
import altair as alt

data =pd.read_csv("cali_wildfires.csv")
data.head()
#source for data: https://www.kaggle.com/ananthu017/california-wildfire-incidents-20132020
```

```python
"# Acres Burned vs different counties in california compared with crews involved in putting ot the fire"
```

```python
alt.Chart(data).mark_circle().encode(
    x='Counties',
    y="AcresBurned",
    color = alt.Color("CrewsInvolved", scale =alt.Scale(scheme = "spectral")),
    size = "Engines"
)

```

```python
"# Notice there is no correlation to Acres Burned, Crews Involved, and Engines but there does seem to be a trend with crews involved and Counties.Lets explore this further!"
```

```python
alt.Chart(data).mark_tick().encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
    color= alt.Color("AcresBurned", scale = alt.Scale(scheme = "spectral")),
    tooltip="Counties"
).properties(
    width=350,
    height=350
).repeat(
    column=['Engines', 'CrewsInvolved', 'AirTankers'],
    row=['AcresBurned']
).interactive()
```

```python
"# Notice how acres burned and different rescources dont have a correlation. Can we find a correlation with diffrent value?"
```

```python
c1=alt.Chart(data).mark_bar().encode(
    x = "Engines",
    y = "Counties",
    tooltip="AcresBurned"
).properties(
    width=350,
    height=500
).interactive()

c2 =alt.Chart(data).mark_bar().encode(
    x = "CrewsInvolved",
    y = "Counties",
    tooltip="AcresBurned"
).properties(
    width=350,
    height=500
).interactive()

c3=alt.Chart(data).mark_bar().encode(
    x = "AirTankers",
    y = "Counties",
    tooltip="AcresBurned"
).properties(
    width=350,
    height=500
).interactive()

c1|c1|c3
```
